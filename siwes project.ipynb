```python
 MATHEMATICAL PREDICTIVE MODELLING AND CORONAVIRUS  PANDEMIC/EPIDEMIC PYTHON PREDICTION
     
  What is Mathematical Modelling?

models describe our beliefs about how the world functions. In mathematical modelling, we translate those beliefs
into the language of mathematics.

A model can also be defined as a prototype (an example) of something that is real. So therefore, A model is a simplified 
idealised representation of a real-life system, which may be in existence already or an idea that is yet to be actualised.


 what objective can modelling achieve?
 
Mathematical modelling can be used for a number of different reasons. but note that how many objective achieved depends
on both the state of knowledge about a system and how well the modelling is done. Examples of the range of objectives.


1. Developing scientific understanding
            
            -through quantitative expresssion of a current knowledge of system (as well as displaying what we know'
             this may also show up what we do not know)

2. To test the effect of changes in a system.

3. To aid decision making' including 
    (i) tactical decisions by managers
    (ii) strategic decisions by planners
    
4. To reduce  complexity and serve as a learning tool.

5. For economic and safety reasons .
 

There are various classifications and types of models, but for this project work we will be focusing on the
predictive type of model.

 WHAT IS A PREDICTIVE MODEL?
 
This model is a model that indicates what the modeled variable will be. it merely lays down the formula to be used
in the neccessary computations.


 STEPS IN MODEL CONSTRUCTION:
 
Given a real-life system and a problem related to it, we may wish to construct a model to help us provide an answer
to the problem.

 THE FOLLOWING STEPS ARE ADAPTED IN THE CONSTRUCTION OF MODELS:
 
 1. DEFINITION OF THE PROBLEM:
     Here, the goal of study is clearly identified and stated, investigate if a similar model has already been developed(consult literature and other people), collect the reqired data, specify the scope and also the limitations of this study etc.
     
2. CONSTRUCTION/FORMULATION OF THE MODEL:
    In this step, the problem defined is translated into mathematical relationships. To do this effectively there is need to specify the variables parameters, constraints, objectives. The model should be as close as a real-life situation as possible in order to make useful decisions  from the resulting solutions.
    
3. SOLUTION OF THE MODEL:
   Here, a procedure is used to derive solutions from the model. The solution method adopted depends on the nature of the model
   formulated. This step also includes sensitivity analysis.
   
4. VALIDATING (TESTING) THE MODEL:
    in this step, we checK to see that the solution does make sense. Does it conform to past performances(historical data) given the similar input conditions? Does the model do what is supposed to do as expected of it? Are the results acceptable?.
    to ensure validity of a model,all the relevant variables of the real-life system must be included in the model and the relationship between the variables must be expressed.
    
    
5. IMPLEMENTATION/APPLICATION OF THE SOLUTION:
   This is the final step. After validating the model to develop an acceptable one, the recommendations are implemented.
```
